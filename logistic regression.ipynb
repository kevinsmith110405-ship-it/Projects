```python
# import library
import pandas as pd
```

```python
# read data
diabetes = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Diabetes.csv')
```

```python
# display first 5 rows
diabetes.head()
```

```python
diabetes.info
```

```python
diabetes.describe()
```

```python
# define y and x
y=diabetes['diabetes']
x=diabetes[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi',
       'dpf', 'age']]
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=2529,train_size=0.7)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# select a model
#from sklearn.linear_model import LogisticRegression
#model = LogisticRegression(max_iter=500)
```

```python
#select an alternative model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
# train the model
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# predict
y_pred=model.predict(x_test)
```

```python
# find the accuracy
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
